```python
from model import SentenceClassifier
import data_transformer
import pandas as pd
import torch
```

```python
model = SentenceClassifier()
model.load_state_dict(torch.load('model.pth'))
model.eval()
```

```python
df = pd.read_table('dataset/test.tsv')
loader = data_transformer.get_loader(df['title'].values, df['is_fake'].values, batch_size=1, shuffle=False)
res = []
for x, _ in loader:
  attention_mask = x['attention_mask']
  input_ids = x['input_ids'].squeeze(1)
  output = model(input_ids, attention_mask).round().int().item()
  res.append(output)
print(res)
```

```python
df['is_fake'] = res
df
```

```python
df.to_csv('predictions.csv', index=False)
```
